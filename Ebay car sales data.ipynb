I have chosen to look at the top 10 of car manufacturers, because they make up over 80% of the used market.

The next step is to calculate the mean prices for each brand

```python
top_10_brands=autos["brand"].value_counts(normalize=True).sort_values(ascending=False).head(10).index
print(top_10_brands)
```

```python
top_10_mean_price_by_brand={}

for brand in list(top_10_brands):
    top_10_mean_price_by_brand[brand]=autos.loc[autos["brand"]==brand,"price"].mean()
    
print(top_10_mean_price_by_brand)
```

```python
#because dicts are orderless, I will transform into a series to perform further analysis.

top_10_mean_price_by_brand_df=pd.DataFrame.from_dict(top_10_mean_price_by_brand,orient="index",)
print(top_10_mean_price_by_brand_df.sort_values(0,ascending=False))
```

On average Audi ist the most expensive brand for used cars, followed by Mercedes and BMW. The gap between those brands is more or less the same.

The next segment is about 3000 Euros away for the first. It consists of Volkswagen, Seat, Ford and Peugeot. Those have an average price gap of about 700 Euros.

The next segment consist of the sub 3000 Euros on average car brands Opel, Fiat and Renault. The average price gaps are between 100 and 350 Euros.

```python
#Write a function for the three date columns

def clean_date(column):
    clean_c=column.str.split(n=1,expand=True)
    clean_c.drop(1,axis=1,inplace=True)
    clean_c=(clean_c[0]
            .str.replace("-","")
            .str.strip()
            .astype(int)
            )
    return clean_c

autos["date_crawled"]=clean_date(autos["date_crawled"])
autos["date_crawled"]
```

```python
autos["last_seen"]=clean_date(autos["last_seen"])
autos["last_seen"]
```

Split the odometer_km into groups, and use aggregation to see if average prices follows any patterns based on the milage.

```python
autos["odometer_km"].describe()
```

```python
autos["odometer_km"].value_counts(normalize=True)
```

I choose to split in five groups. Up to 30,000km, up to 60,000km and so forth, till 150,000km

```python
mileage_groups=[30000,60000,90000,150000]
avg_price_m_group={}

for group in mileage_groups:
    price_only_mileage=autos.loc[autos["odometer_km"].between(group-30000,group),"price"]
    avg_price_m_group[group]=price_only_mileage.mean()
    
mileage_price=pd.DataFrame.from_dict(avg_price_m_group,orient="index")
mileage_price.columns=["avg_price"]
mileage_price
```

There is a correspondance between mileage and the average price.
The biggest drops start at over 60,000 kilometers (about 2/3 of the price of cars with 30,000 kilometers or less). Cars with 150,000 kilometers or more are especially cheap (about 1/4).

How much cheaper are cars with damage than their non-damaged counterparts?

```python
#I start off with calculating the means for every model, damaged and non-damages
damaged_non_damaged=pd.DataFrame(columns=["avg_price_undamaged","avg_price_damaged"])
d_nd_values=[]


for model in autos["model"].unique():
    only_model=autos.loc[autos["model"]==model]
    d_nd_values.append(only_model.loc[autos["unrepaired_damage"]=="nein","price"].mean())
    d_nd_values.append(only_model.loc[autos["unrepaired_damage"]=="ja","price"].mean())
    damaged_non_damaged.loc[model]=d_nd_values
    d_nd_values.clear()

    
#No i have to calculate difference for each model and calculate a mean between all differences
mean_difference_d_nd=(damaged_non_damaged["avg_price_undamaged"]-damaged_non_damaged["avg_price_damaged"]).mean()
print(damaged_non_damaged)
print(mean_difference_d_nd)
```

The average difference between a damaged use car and a non-damaged used car (between the same car models) is about 3807 Euro. Of course this is just a rough estimation, other variables have to be controlled to determine their influence (e.g. the mileage, etc.).
